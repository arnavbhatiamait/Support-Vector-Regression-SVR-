Support Vector Regression (SVR)

Importing The Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing Data Set

```python
df=pd.read_csv('Position_Salaries.csv')
df
```

```python
x=df.iloc[:,1:2].values
x
```

```python
y=df.iloc[:,-1].values
y
y=y.reshape(len(y),1)
print(y)
```

Features Scaling 

```python
#! salaries are way higher than than the level 
# todo Apply Features Scaling
# * values from -3 to 3
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
sc_y=StandardScaler()
y=sc_y.fit_transform(y)
x=sc_x.fit_transform(x)
```

```python
x
```

```python
y
```

Training The SVR Model in the whole DataSet

```python
from sklearn.svm import SVR
SVR_regressor=SVR(kernel='rbf')
#!rbf = Radial Basis Function Kernel
SVR_regressor.fit(x,y)
```

Prediction of A New Result

```python
sc_y.inverse_transform(SVR_regressor.predict(sc_x.transform([[6.5]])).reshape(-1,1))
```

Visualize The SVR Results

```python
plt.scatter(sc_x.inverse_transform(x),sc_y.inverse_transform(y),color='red')
plt.plot(sc_x.inverse_transform(x),sc_y.inverse_transform(SVR_regressor.predict(x).reshape(-1,1)),color='blue')
plt.title("Results using the SVR")
plt.xlabel("Possition Level")
plt.ylabel("Salary")
plt.show()
```

Visualize The SVR Results In High Resolution And Accuracy

```python
x_grid=np.arange(min(sc_x.inverse_transform(x)),max(sc_x.inverse_transform(x)),0.1)
x_grid=x_grid.reshape(len(x_grid),1)
plt.scatter(sc_x.inverse_transform(x),sc_y.inverse_transform(y),color='red')
plt.plot(x_grid,sc_y.inverse_transform(SVR_regressor.predict(sc_x.transform(x_grid)).reshape(-1,1)),color='blue')
plt.title("Results using the SVR High Accuracy")
plt.xlabel("Possition Level")
plt.ylabel("Salary")
plt.show()
```
